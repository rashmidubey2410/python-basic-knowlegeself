```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0.0,2.0,0.01)
s= 1+np.cos(5*np.pi*t)
plt.plot(t,s)
plt.show()
```

```python
t=np.arange(0.0,2.0,0.01)
s= np.cos(5*np.pi*t)
plt.plot(t,s)
plt.show()
```

```python
t=np.arange(0.0,2.0,0.01)
s= 1+np.sin(5*np.pi*t)
plt.plot(t,s,'r')
plt.grid()
plt.show()
```

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'r')
plt.show()

plt.subplot(2,1,2)
plt.plot(x2,y2,'g')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[12,23,27,11,21]
tick=['one','two','three','four','five']

plt.bar(x,y,tick_label=tick,width=0.5,color=['m','g'])

plt.xlabel('num')
plt.ylabel('hight')
plt.title('bar chart')
plt.grid()

plt.show()
```

```python
#histograf

age=[5,44,23,21,4,23,35,45,70,56,54,62,22,11,1,5,30,77,50,57,44,99]
range=(0,100)
bins = 10

plt.hist(age,bins,range,color='g',histtype='bar',rwidth=0.5)

plt.show()
```

```python
x= [1,2,3,4,5,3,7,8,9,10]
y= [2,1,4.5,3,2,4,7,2,2,4,]

plt.scatter(x,y,label='pollstar',color='r',marker='*',s=66)

plt.legend()
plt.show()
```

```python
act=['eat','sleep','play','cry','walk']
slices=[3,4,2,1,6]
color=['g','y','b','r','m']

plt.pie(slices,labels=act,colors=color,startangle=90,shadow=True,
       explode=(0,0,0.2,0,0),autopct='%1.2f%%')
plt.legend()
plt.show()
```

```python
x= np.arange(0,2*np.pi,0.01)

y1= np.sin(x)
y2= np.cos(x)

plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sincos')
plt.legend()


plt.show()
```
