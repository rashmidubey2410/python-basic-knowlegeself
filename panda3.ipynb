```python
#rankmethod


import pandas as pd
nba7=pd.read_csv("nba.csv")
nba7.head()
```

```python
import pandas as pd
nba7=pd.read_csv("nba.csv").dropna(how="all")

nba7["Salary"]=nba7["Salary"].fillna(0)
nba7.head()
```

```python
nba7["Salaryrank"]=nba7["Salary"].rank(ascending=False).astype(int)
nba7.head()
```

```python
nba7["Salaryrank"]=nba7["Salary"].rank(ascending=False).astype(int)

nba7.sort_values(by="Salary")

nba7.head()
```
